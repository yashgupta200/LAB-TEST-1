```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import butter, filtfilt

np.random.seed(42)
minutes = 1440
base_traffic = np.random.poisson(lam=50, size=minutes)
noise = np.random.normal(loc=0, scale=10, size=minutes)
vehicle_count = base_traffic + noise
vehicle_count = np.clip(vehicle_count, 0, None)

def low_pass_filter(data, cutoff=0.1, fs=1.0, order=5):
    nyquist = 0.5 * fs
    normal_cutoff = cutoff / nyquist
    b, a = butter(order, normal_cutoff, btype='low', analog=False)
    return filtfilt(b, a, data)

smoothed_data = low_pass_filter(vehicle_count)

hourly_average = np.array([np.mean(vehicle_count[i:i+60]) for i in range(0, minutes, 60)])

plt.figure(figsize=(15, 8))
plt.plot(vehicle_count, label='Noisy Data', color='gray', alpha=0.5)
plt.plot(smoothed_data, label='Smoothed Data', color='blue')
plt.scatter(np.arange(0, minutes, 60), hourly_average, color='red', marker='o', label='Hourly Averages')
plt.title('Traffic Flow Monitoring')
plt.xlabel('Minutes')
plt.ylabel('Number of Vehicles')
plt.legend()
plt.grid()
plt.show()

def detect_congestion(data, threshold=120, duration=15):
    congestion_intervals = []
    count = 0
    for i in range(len(data)):
        if data[i] > threshold:
            count += 1
            if count == duration:
                start = i - duration + 1
                congestion_intervals.append((start, i))
        else:
            count = 0
    return congestion_intervals

congestion_intervals = detect_congestion(vehicle_count)

for start, end in congestion_intervals:
    plt.axvspan(start, end, color='red', alpha=0.3)

plt.show()
```
